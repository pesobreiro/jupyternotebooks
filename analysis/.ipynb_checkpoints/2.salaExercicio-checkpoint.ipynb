# Dados recolhidos sala de exercício Alv

```python
import pandas as pd
import seaborn as sns

dfSEAlv = pd.read_excel('../dados/Elitefitness Alverca - Sala exercício(1-214).xlsx',sheet_name='Sheet1')
```

```python
dfSEAlv.columns
```

```python
dfSEAlv.head()
```

```python
cols = ['ID', 'E-mail', 'Nome', 'Pontuação total', 'Feedback do questionário',  'Pontos – Data',
        'Feedback – Data', 'Pontos – Hora', 'Feedback – Hora','Pontos – Número pessoas sexo feminino','Feedback – Número pessoas sexo feminino',
        'Pontos – Número pessoas sexo masculino','Feedback – Número pessoas sexo masculino', 'Pontos – Quantas pessoas cárdio?',
        'Feedback – Quantas pessoas cárdio?','Pontos – Quantas pessoas a fazerem abdominal?','Feedback – Quantas pessoas a fazerem abdominal?',
        'Pontos – Quantas pessoas ombros?','Feedback – Quantas pessoas ombros?', 'Pontos – Quantas pessoas peito?', 
        'Feedback – Quantas pessoas peito?','Pontos – Quantas pessoas costas?','Feedback – Quantas pessoas costas?',
        'Pontos – Quantas pessoas treino de pernas?','Feedback – Quantas pessoas treino de pernas?']
dfSEAlv.drop(columns=cols, index=1, inplace=True)
```

```python
dfSEAlv.columns
```

```python
dfSEAlv.columns=['horaInicio','horaConclusao','data','hora','mulheres','homens','cardio','abdominal','ombros','peito','costas','pernas']
```

```python
dfSEAlv.head()
```

```python
dfSEAlv.info()
```

```python
dfSEAlv.fillna(0,inplace=True)
```

```python
dfSEAlv['mulheres']=dfSEAlv['mulheres'].apply(lambda x:str(x).replace('"','0'))
```

```python
dfSEAlv.loc[dfSEAlv['mulheres'].str.isalpha(),'mulheres'] = 0
```

```python
dfSEAlv.mulheres.unique()
```

```python
dfSEAlv.head()
```

```python
dfSEAlv.mulheres = dfSEAlv.mulheres.astype(int)
```

```python
dfSEAlv['total']=dfSEAlv.mulheres+dfSEAlv.homens
```

```python
dfSEAlv.head()
```

## Colocar tudo em ints

```python
cols = ['mulheres', 'homens','cardio', 'abdominal', 'ombros', 'peito', 'costas', 'pernas', 'total']
for col in cols:
    dfSEAlv[col] = dfSEAlv[col].astype(int)
```

```python
dfSEAlv.columns
```

## Converter hora para tempo

```python
dfSEAlv['hora'] = pd.to_datetime(dfSEAlv['hora'])
```

## Quantas pessoas estão nas sala de exercício

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [12, 7]
```

```python
dfSEAlv.hist();
```

```python
plt.rcParams['figure.figsize'] = [12, 5]
dfSEAlv.boxplot();
```

### Quantas pessoas?

### Pessoas

```python
dfSEAlv['horaSo']=dfSEAlv['hora'].apply(lambda x: x.hour)
dfSEAlv.horaSo.unique()
```

```python
plt.rcParams['figure.figsize'] = [15, 10]
dfSEAlv[['horaSo','mulheres', 'homens','cardio', 'abdominal', 'ombros', 'peito', 'costas', 'pernas', 'total']].boxplot(by='horaSo');
```

```python
dfSEAlv.columns
```

```python
cols = ['mulheres', 'homens','cardio', 'abdominal', 'ombros', 'peito', 'costas', 'pernas', 'total']
```

```python
plt.rcParams['figure.figsize'] = [15, 5]
for col in cols:
    plt.figure()
    sns.boxplot(x='horaSo',y=col,data=dfSEAlv,palette='rainbow').set_title(col +' sala exercício Alverca');
```

O que é que se passa aqui? Ninguém faz abdominais????

## Resumo num boxplot

```python
dfSEAlv.columns
```

```python
dfSEAlv.head()
```

```python
dfMelt=pd.melt(dfSEAlv,id_vars=['horaSo'],value_vars=cols,var_name='exercicio')
```

```python
dfMelt
```

```python
plt.rcParams['figure.figsize'] = [20, 7]
```

```python
sns.boxplot(x='horaSo',y='value',hue='exercicio',data=dfMelt,palette='rainbow').set_title('Sala de exercicio');
```
